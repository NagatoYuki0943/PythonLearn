# putpalette 给图片添加调色板

```python
def colors(num_colors):
    hsvs = [(n / num_colors, 0.8, 1.) for n in range(num_colors)]
    random.shuffle(hsvs) # 随机打乱颜色顺序
    rgbs = [colorsys.hsv_to_rgb(*hsv) for hsv in hsvs]
    rgbs = [(int(rgb[0] * 255), int(rgb[1] * 255), int(rgb[2] * 255)) for rgb in rgbs]
    return rgbs
```

```python
palette = colors(256 - 1)
palette
```

```python
palette.insert(0, (0, 0, 0))
palette = list(np.array(np.array(palette)).reshape(-1))
len(palette)
```

```python
image_restore.getpalette()
```

```python
image_restore.putpalette(palette)
```
