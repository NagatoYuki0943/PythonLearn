```python
import os
import sys
from matplotlib import pyplot as plt

from PIL import Image, ImageOps
```

# 打开图片

```python
image = Image.open('./image/71479714_p0.jpg')

print(image)
```

# 图像有时会自动旋转，这样可以保证角度相同

```python
# 恢复正常角度的图像
image = ImageOps.exif_transpose(image)
```

# 显示图片

```python
image
```

```python
image.show()
```

# 保存图片,可以更改格式

```python
image.save('./image/71479714_p0.png')
```

# plt画图

```python
plt.figure(figsize=(10,8))
plt.axis('off') # 不显示坐标轴
plt.title('黄前久美子')
plt.imshow(image)
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
plt.show()
```

```python
print(type(image))
print(image.size)  				# 图片的尺寸 width height
print(image.height)             # 1654
print(image.width)              # 1048
print(image.mode)  				# 图片的模式
print(image.format)  			# 图片的格式
```
