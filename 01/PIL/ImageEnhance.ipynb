```python
import numpy as np
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
image: Image = Image.open("squirrel.jpg")
image
```

# ImageEnhance.Brightness 调整亮度

```python
enhancer = ImageEnhance.Brightness(image)
enhancer
```

```python
enhancer.enhance(0.5)
```

```python
enhancer.enhance(2)
```

# ImageEnhance.Contrast 调整对比度

```python
enhancer = ImageEnhance.Contrast(image)
enhancer
```

```python
enhancer.enhance(0.5)
```

```python
enhancer.enhance(2)
```

# ImageEnhance.Sharpness 调整锐度

```python
enhancer = ImageEnhance.Sharpness(image)
enhancer
```

```python
enhancer.enhance(0.5)
```

```python
enhancer.enhance(2)
```

# ImageEnhance.Color 调整颜色

```python
enhancer = ImageEnhance.Color(image)
enhancer
```

```python
enhancer.enhance(0.5)
```

```python
enhancer.enhance(2)
```
