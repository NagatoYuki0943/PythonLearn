## 添加噪声

```python
image = Image.open('./image/71479714_p0.jpg')
img = np.array(image)

#随机生成5000个椒盐
rows, cols, dims = img.shape
for i in range(5000):
    x=np.random.randint(0,rows)
    y=np.random.randint(0,cols)
    img[x,y,:]=255
    
plt.figure(figsize=(9, 16))
plt.imshow(img)
plt.axis('off')
plt.show()
```

# 二值化处理

```python
image = Image.open('./image/71479714_p0.jpg')
#                 黑白图片
image = image.convert('L')
img = np.array(image)
img
```

```python
rows,cols=img.shape
for i in range(rows):
    for j in range(cols):
        if (img[i,j]<=128):
            img[i,j]=0
        else:
            img[i,j]=1
img
```

```python
plt.figure(figsize=(9, 16))
#         直接使用numpy数组就能输出
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```
