# 创建一个新的图片 Image.new()

参数:

- mode: "1", "CMYK", "F", "HSV", "I", "L", "LAB", "P", "RGB", "RGBA", "RGBX", "YCbCr"
- size: (宽, 高)
- color: float / 文字如'red' / (0~255, 0~255, 0~255)

```python
image1 = Image.new('RGB', (900, 600), 100)
plt.figure(figsize=(9, 6))
plt.imshow(image1)
plt.show()
```

```python
image2 = Image.new('RGB', (900, 600), 'red')
plt.figure(figsize=(9, 6))
plt.imshow(image2)
plt.axis('off')
plt.show()
```

```python
image3 = Image.new('RGB', (900, 600), (0, 255, 0))
plt.figure(figsize=(9, 6))
plt.imshow(image3)
plt.axis('off')
plt.show()
```

# 层叠图片 Image.blend() 
层叠两个图片，image1和image2, 

alpha是一个介于[0,1]的浮点数

- 如果为0，  效果为image1

- 如果为1.0，效果为image2

当然image1和image2的尺寸和模式必须相同。这个函数可以做出很漂亮的效果来，而图形的算术加减后边会说到。

## 如果为1.0，效果为image2

```python
blend1 = Image.blend(image1, image2, 1) 
plt.figure(figsize=(9, 6))
plt.imshow(blend1)
plt.show()
```

## 如果为0，效果为image1

```python
blend2 = Image.blend(image1, image2, 0) 
plt.figure(figsize=(9, 6))
plt.imshow(blend2)
plt.show()
```

## 0.5, 两者混合

```python
blend3 = Image.blend(image1, image2, 0.5) 
plt.figure(figsize=(9, 6))
plt.imshow(blend3)
plt.show()
```

# 蒙板 Image.composite(image1, image2, mask)
composite可以使用另外一个图片作为蒙板(mask)，所有的这三张图片必须具备相同的尺寸，mask图片的模式可以为“1”，“L”，“RGBA”

```python
mask = Image.new('1',(900, 600))
```

```python
composite = Image.composite(image3, image3, mask) 
plt.figure(figsize=(9, 6))
plt.imshow(composite)
plt.show()
```
